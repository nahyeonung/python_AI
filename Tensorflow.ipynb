```python
import tensorflow as tf

#tf.constant: consant는 상수를 뜻한다. 상수 텐서를 만들 때 사용한다.
tensor = tf.constant(3)
tensor2 = tf.constant([[1,2],[3,4]]) #대괄호를 해야 shape이 잡힌다.
print(tensor)
print(tensor2)

#tf.zeros(): 괄호안의 수 만큼 0을 생성해준다.
tensor3 = tf.zeros([2,2,3]) # 2행 3열인 행렬을 2개 생성해준다.(여기서 맨 앞의 2가 두 개 생성의 의미)
print(tensor3)
```

```python
#tf.Variable: Variable은 변수를 뜻한다. 수학에서는 변수가 변하는 숫자이지만 텐서플로에서는 Constant와 큰 차이가 없다
#딥러닝에서 Variable은 weight값을 저장할 때 쓰인다. 즉 딥러닝 과정에서 연속적으로 연산을 해야할 때 사용.
#ex) w = w + 1 즉 값을 업데이트 하고 싶을 때 쓰이는 값
v1 = tf.Variable(1)
v2 = tf.Variable([1,2])
print(v1)
print(v2)
```

```python
#tensorflow의 장점. 행렬 연산이 쉽다.
tensor = tf.constant([[3,4,5],
                      [1,2,3]])
tensor2 = tf.constant([[6,7,8],
                       [3,4,5]])
print(tensor + tensor2)
print(tf.add(tensor, tensor2))
```

```python
import cv2
import matplotlib.pyplot as plt

#원소 간 곱 VS 행렬 곱
#원소 간 곱은 흔히 아는 원소 끼리 곱하는 단순 방식.
#행렬 곱은 하나의 행렬의 가로와 다른 하나의 행렬의 세로의 원소를 각각 곱해서 더해주는 것

tensor = tf.constant([[1,2],
                      [3,4]])
tensor2 = tf.constant([[5,6],
                      [7,8]])

print(tf.math.multiply(tensor, tensor2)) #원소간 곲
print(tf.matmul(tensor, tensor2)) # 행렬 곱

img = cv2.imread('matrix_multiply.png', cv2.IMREAD_COLOR)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
     
```
