```python
import time
import cv2
import matplotlib.pyplot as plt #jupyter     
# cv2  matplotlib   RGB    .
# OpenCV: [B, G, R] / Matplotlib: [R, G, B]
# cv2     , matplotlib    .

# cv2   
# img_basic = cv2.imread('dog.jpg', cv2.IMREAD_COLOR)
# cv2.imshow('Img',img_basic)
# cv2.waitKey(0)
# cv2.imwrite('result1.jpg', img_basic)

# img_gray = cv2.cvtColor(img_basic, cv2.COLOR_BGR2GRAY)
# cv2.imshow('png',img_gray)
# cv2.waitKey(0)
# cv2.imwrite('result2.jpg', img_gray)

# matplotlib 
img_basic = cv2.imread('dog.jpg', cv2.IMREAD_COLOR)
plt.imshow(cv2.cvtColor(img_basic, cv2.COLOR_BGR2RGB)) #COLOR_BGR2RGB: BGR RGB 
plt.show()

img_basic = cv2.cvtColor(img_basic, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(img_basic, cv2.COLOR_GRAY2RGB))
plt.show()
```

```python
#   
image = cv2.imread('dog.jpg')

start_time = time.time()
for i in range(0, 100):
    for j in range(0, 100):
        image[i, j] = [255, 255, 255] #255,255,255   
print("--- %s seconds ---" % (time.time() - start_time)) #print   - start_time  

start_time = time.time()
image[0:100, 0:100] = [0, 0, 0] #0,0,0   .   for    
print(" --- %s seconds ---" % (time.time() - start_time))

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
#ROI(Reason Of Interest)  

image = cv2.imread('dog.jpg')

#Numpy Slicing: ROI  
roi = image[100:250, 50:200]

# #ROI   
image[0:150, 0:150] = roi #[150,150] [150,150]    .      !

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
image = cv2.imread('cat.jpg')
image_ = image[0:250, 0:250] 
image2 = cv2.imread('dog.jpg')
image2_ = image2[0:250, 0:250] #cat dog       
print(image_.shape)
print(image2_.shape)

result = cv2.add(image_,image2_)
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.show()

result = image_ + image2_
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
#GRAY COLOR      !
image = cv2.imread('cat.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
image_ = image[0:250, 0:250] 
image2 = cv2.imread('dog.jpg')
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)

image2_ = image2[0:250, 0:250] #cat dog       
print(image_.shape)
print(image2_.shape)

result = cv2.add(image_,image2_)
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.show()

result = image_ + image2_
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.show()
```
