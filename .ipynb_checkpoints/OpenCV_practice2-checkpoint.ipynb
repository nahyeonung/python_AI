```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('dog.jpg', cv2.IMREAD_GRAYSCALE)

images = []
ret, thres1 = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
ret, thres2 = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY_INV)
ret, thres3 = cv2.threshold(image, 127, 255, cv2.THRESH_TRUNC)
ret, thres4 = cv2.threshold(image, 127, 255, cv2.THRESH_TOZERO)
ret, thres5 = cv2.threshold(image, 127, 255, cv2.THRESH_TOZERO_INV)
images.append(thres1)
images.append(thres2)
images.append(thres3)
images.append(thres4)
images.append(thres5) #LIST에 이미지를 넣고 FOR문으로 하나씩 다 출력

for i in images: 
    plt.imshow(cv2.cvtColor(i, cv2.COLOR_GRAY2RGB))
    plt.show()
```

```python
#Gaussian으로 임계점 처리하기
image = cv2.imread('dog.jpg', cv2.IMREAD_GRAYSCALE)

ret, thres1 = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
thres2 = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 21, 3)
thres3 = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 21, 3)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_GRAY2RGB))
plt.show()

plt.imshow(cv2.cvtColor(thres1, cv2.COLOR_GRAY2RGB))
plt.show()

plt.imshow(cv2.cvtColor(thres2, cv2.COLOR_GRAY2RGB))
plt.show()

plt.imshow(cv2.cvtColor(thres3, cv2.COLOR_GRAY2RGB))
plt.show()
```

```python
import cv2
import numpy as np

# cv2.createTrackbar(trackbarname, windowname, start, end, onChange):트랙바를 지정된 윈도우에 생성하는 함수
# start:트랙바 시작 값
# end: 트랙바 끝 값
# onChange: 트랙바 이벤트 발생시 수행되는 콜백 함수

# cv2.getTrackbarPos(trackbarname, windowname): 트랙바의 현재 위치를 리턴하는 함수




def change_color(x):
    r = cv2.getTrackbarPos("R", "Hyoenung")
    g = cv2.getTrackbarPos("G", "Hyoenung")
    b = cv2.getTrackbarPos("B", "Hyoenung")
    image[:] = [b, g, r] #리스트 전체에 대해서 b,g,r순서대로 값을 더하겠다
    cv2.imshow("Hyoenung", image)

image = np.zeros((600, 400, 3), np.uint8) #600x400크기의 이미지를 생성
cv2.namedWindow("Hyoenung")

cv2.createTrackbar("R", "Hyoenung", 0, 255, change_color)
cv2.createTrackbar("G", "Hyoenung", 0, 255, change_color)
cv2.createTrackbar("B", "Hyoenung", 0, 255, change_color)

cv2.imshow("Hyeonung", image)
cv2.waitKey(0)
```
