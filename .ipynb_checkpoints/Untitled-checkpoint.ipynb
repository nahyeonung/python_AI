```python
# 라이브러리 임포트하기

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='darkgrid')

%matplotlib inline
```

```python
# 데이터셋
tips = sns.load_dataset('tips') #seaborn에서 사용하는 기본적이 데이터셋 중에 tips에 관련된 데이터셋을 불러옴
tips
```

```python
sns.relplot(x='total_bill', y='tip', data=tips) #x축을 total_bill, y축을 tip로 지정하고 relplot형태로 출력
```

```python
#hue 원하는 시리즈를 기준으로 색상을 달리하는 속성
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker') #흡연자에 따라서 색상을 달리 출력
```

```python
#style $ hue 모양과 색상으로 종류를 더 많이 구별하기
sns.relplot(x='total_bill', y='tip', data=tips, style='smoker', hue='size')
```

```python
#색상 정하기
sns.relplot(x="total_bill", y="tip", data=tips, style="size", hue="smoker", palette="Blues")
```
